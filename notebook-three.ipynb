```python
import json
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/asteroid_close_approaches_2015_2035.csv')
df_clean = df[['designation', 'dist_km', 'velocity_km_s', 'close_approach_date']].dropna()

asteroids_list = df_clean.to_dict(orient='records')

#Data saving in .json
with open ('./data/asteroids.json', 'w') as f:
    json.dump(asteroids_list, f, indent=4)

print("JSON done!", len(asteroids_list))
```
